```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd


def n():
    print('-'*50, '\n')

def parse_sonar_file(path):
    sonar_dict = {}
    current_key = None
    current_values = []

    with open(path, 'r') as f:
        for line in f:
            line = line.strip()

            # 1) KEY line
            if line.startswith("*"):
                current_key = line.replace(":", "")
                current_values = []
                continue

            # 2) Opening of values block {
            if line.startswith("{"):
                # Remove the "{" before splitting values
                line = line[1:]

            # 3) Closing of values block }
            if line.endswith("}"):
                # Remove "}" before processing
                line = line[:-1]

                # Add last line of values
                if line.strip():
                    numbers = [float(x) for x in line.split()]
                    current_values.extend(numbers)

                # Save to dictionary
                sonar_dict[current_key] = current_values

                # Reset for next block
                current_key = None
                current_values = []
                continue

            # 4) Regular lines of values
            if current_key is not None:
                if line.strip():
                    numbers = [float(x) for x in line.split()]
                    current_values.extend(numbers)

    return sonar_dict
```
